```python
from core.apk_util import *


def read_transitions(file_path):
    transitions = {}
    temp_transitions = []
    with open(file_path) as f:
        lines = f.readlines()
        for line in lines:
            if "APK: " in line:
                continue
            if "Tool: " in line:
                tool_name = line.split("Tool: ")[-1].strip()
                if tool_name not in transitions:
                    transitions[tool_name] = []
                transitions[tool_name].extend(temp_transitions)
                temp_transitions = []
            if "->" in line:
                src, tgt = line.strip().split(" -> ")
                if "Start" in src:
                    continue
                # src, tgt = transform_activtiy_name(src, package_name), transform_activtiy_name(tgt, package_name)
                transition = f"{src} -> {tgt}"
                temp_transitions.append(transition)
    return transitions
```

```python
result_dirs = [
    "first_round_no_internet",
    "second_round_no_internet",
    "mars_first_round_no_internet",
    # "mars_second_round_no_internet"
]

from core.file_paths import *

all_results = {}
for result_dir in result_dirs:
    print(f"reading transitions for {result_dir}")
    file_path = f"summary/data_analysis/{result_dir}/all_raw_transitions.txt"
    round_transitions = read_transitions(file_path)
    for tool in tools:
        if tool not in all_results:
            all_results[tool] = {}
        all_results[tool][result_dir] = round_transitions[tool]
```

```python
for tool in tools:
    print(tool)
    f1_saturn = all_results[tool]["first_round_no_internet"]
    f1_mars = all_results[tool]["mars_first_round_no_internet"]
    f2_saturn = all_results[tool]["second_round_no_internet"]
    print(f"first round no internet: {len(f1_saturn)}")
    print(f"mars first round no internet: {len(f1_mars)}")
    print(f"second round no internet: {len(f2_saturn)}")
    union = set(f1_saturn).union(set(f1_mars)).union(set(f2_saturn))
    print(f"union: {len(union)}")
    f1_saturn_unique = set(f1_saturn).difference(set(f1_mars).union(f2_saturn))
    f2_saturn_unique = set(f2_saturn).difference(set(f1_mars).union(f1_saturn))
    f1_mars_unique = set(f1_mars).difference(set(f1_saturn).union(f2_saturn))
    print(f"unique in f1_saturn: {len(f1_saturn_unique)}")
    print(f"unique in f1_mars: {len(f1_mars_unique)}")
    print(f"unique in f2_saturn: {len(f2_saturn_unique)}")
```
